```python
from functools import lru_cache
@lru_cache(maxsize=None) # отменяем сохранение кэша
def f(n):
    if n<3:
        return 1
    elif n>2 and n%2 == 0:
        return g(n)+f(n-1)
    elif n>2 and n%2 != 0:
        return f(n-2) -2*g(n+1)
@lru_cache(maxsize=None) # отменяем сохранение кэша
def g(n):
    if n<3:
        return 1
    elif n>2 and n%2 == 0:
        return f(n-3)+f(n-2)
    elif n>2 and n%2 != 0:
        return f(n+1) - g(n-1)
print(g(120))
```

```python
# Алгоритм вычисления значения функции F(n) и G(n), где
# nn – целое число, задан следующими соотношениями:
# F(n)=G(n−1)−G(n−5)
# G(n)=2×(n+1), n≤25
# G(n)=G(n−4)+n, если n>25
# Чему равно значение выражения F(150774)?
from functools import lru_cache
@lru_cache(maxsize=None) # отменяем сохранение кэша
def f(n):
    return g(n-1) - g(n-5)
@lru_cache(maxsize=None) # отменяем сохранение кэша
def g(n):
    if n <= 25:
        return 2*(n+1)
    elif n>25:
        return g(n-4)+n
c = 0
for i in range(150775):
    if f(i) >= 4242:
        c = f((i))
print(c)
```

```python
# Алгоритм вычисления значения функции F(n), где n - целое неотрицательное число, задан следующими соотношениями:
# F(n)=0, при n≤1,
# F(n)=n+F(n/6−2), когда n>1 и кратно 6,
# F(n)=n+F(n+6), когда n>1 и не кратно 6.
# Чему равно минимальное значение n, для которого F(n) определено и превосходит 4242?
from functools import lru_cache
@lru_cache(maxsize=None) # отменяем сохранение кэша
def f(n):
    if n > 1 and n%6==0:
        return n+f(n/6-2)
    elif n>1:
        return n+f(n+6)
    else:
        return 0
# анализируя рекурсию, мы видим, что при n>1 и не кратно 6, рекурсия будет бесконечной
# функция определена - имеет конечное решение. Значит будем отталкиваться от условия n>1 и кратно 6
# при n = 6 получаем 6
# при n = 12 получаем 12
# при n = 18 получаем 18
# складывается впечатление, что ответ близок: берем числа, кратные 6, но нет.
# при n = 24 получается бесконечность
# какое основное условие? Основное условие конечного результата реккурсии - n/6-2 
q = [6,12,18]
for i in range(24,100000,6):
    if i/6-2 in q: # n/6-2 , берем только такие n, при которых получаем 6,12,18
        q.append(i) # постепенно, находя n, пополняем список для наглядности
print(q) # список n дающих конечную рекурсию
w = [] # для наглядности создам список результатов рекурсий 
for i in q:
    w.append(f'f({i})={f(i)}')
print(w) # выбираем n = 4404 , потому что f(4404)=5274.0 > 4242
```
