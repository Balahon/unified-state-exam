```python
# открываем файл на чтение через пробел
zadanie17 = open('17.txt').read().split() 
#print(zadanie17)
# запишем в виде списка, представив все элементы числовами
s1 =list(map(int,zadanie17))
#print(s1)
count = 0 #счётчик
max = 0 #макс
for i in range(len(s1)):
    if i == 0: # 0 элемент не считаем
        next #пропусаем
    if (s1[i-1] + s1[i]) % 5 == 0 and (s1[i-1] % 3 == 0 or s1[i] % 3 == 0):
        count += 1
        if (s1[i-1] + s1[i]) > max:
            max = s1[i-1] + s1[i]
print(count, max)
```

```python
# Дана программа для редактора:
# НАЧАЛО
# ПОКА нашлось (25) ИЛИ нашлось (355) ИЛИ нашлось (555)
#  ЕСЛИ нашлось (25)
#  ТО заменить (25, 5)
#  КОНЕЦ ЕСЛИ
#  ЕСЛИ нашлось (355)
#  ТО заменить (355, 52)
#  КОНЕЦ ЕСЛИ
#  ЕСЛИ нашлось (555)
#  ТО заменить (555, 3)
#  КОНЕЦ ЕСЛИ
# КОНЕЦ ПОКА
# КОНЕЦ
# На вход приведённой выше программе поступает строка, начинающаяся с цифры "2", а затем содержащая n цифр "5" (n > 3).
# Определите наименьшее значение n, при котором в строке, получившейся в результате выполнения программы, количество цифр «3» равно 2.
for x in range(3,50):
    c = "2"+"5"*x
    while("25" in c or "355" in c or "555" in c):
        c = c.replace('25','5',1)
        c = c.replace('355','52',1)
        c = c.replace('555','3',1)
    if c.count("3") == 2:
        print(x)
        print(c)
        break
```
