```python
# Егор составляет таблицу кодовых слов для передачи сообщений. 
# Каждому из них соответствует своё 4-буквенное кодовое слово из букв Б, Л, О, Г, Е, Р. 
# Буква Г используется в слове ровно 1 раз, а гласные не могут стоять рядом. 
# Каждая из других допустимых букв может встречаться в кодовом слове любое количество раз или не встречаться совсем.
# Сколько различных кодовых слов может использовать Егор?
slovo = 'БЛОГЕР'
c = 0
c1 = 0
for i1 in slovo:
    for i2 in slovo:
        for i3 in slovo:
            for i4 in slovo:
                q = i1+i2+i3+i4 # все слова из букв БЛОГЕР
                if (q.count('Г')) == 1: 
                    c += 1 # считаем все слова, что имеют одну букву Г
                    if ((i1 in "ОЕ") and (i2 in "ОЕ")) or ((i2 in "ОЕ") and (i3 in "ОЕ")) or ((i3 in "ОЕ") and (i4 in "ОЕ")):
                        c1 += 1 # считаем все слова, в которых 1 буква Г и глассные буквы стоят рядом
print(c-c1) # из всех слов, что имеют одну букву Г отнимаем все слова в которых 1 буква Г и глассные буквы стоят рядом
# таким образом получим все слова что имеют одну букву Г и глассные не стоят рядом
```

```python
# Все шестибуквенные слова, составленные из букв К, Ц, И, Ч, П записаны в алфавитном порядке и пронумерованы начиная с 1. 
# Вот начало этого списка:
# ИИИИИИ
# ИИИИИК
# ИИИИИП
# ИИИИИЦ
# ИИИИИЧ
# ИИИИКИ
# Сколько слов в списке содержат ровно три буквы И?
sl = 'ИКПЦЧ'
c = 0
for i1 in sl:
    for i2 in sl:
        for i3 in sl:
            for i4 in sl:
                for i5 in sl:
                    for i6 in sl:
                        q = i1+i2+i3+i4+i5+i6
                        if q.count('И') == 3:
                            c+=1
print(c)
```
