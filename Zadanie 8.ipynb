```python
# Егор составляет таблицу кодовых слов для передачи сообщений. 
# Каждому из них соответствует своё 4-буквенное кодовое слово из букв Б, Л, О, Г, Е, Р. 
# Буква Г используется в слове ровно 1 раз, а гласные не могут стоять рядом. 
# Каждая из других допустимых букв может встречаться в кодовом слове любое количество раз или не встречаться совсем.
# Сколько различных кодовых слов может использовать Егор?
slovo = 'БЛОГЕР'
c = 0
c1 = 0
for i1 in slovo:
    for i2 in slovo:
        for i3 in slovo:
            for i4 in slovo:
                q = i1+i2+i3+i4 # все слова из букв БЛОГЕР
                if (q.count('Г')) == 1: 
                    c += 1 # считаем все слова, что имеют одну букву Г
                    if ((i1 in "ОЕ") and (i2 in "ОЕ")) or ((i2 in "ОЕ") and (i3 in "ОЕ")) or ((i3 in "ОЕ") and (i4 in "ОЕ")):
                        c1 += 1 # считаем все слова, в которых 1 буква Г и глассные буквы стоят рядом
print(c-c1) # из всех слов, что имеют одну букву Г отнимаем все слова в которых 1 буква Г и глассные буквы стоят рядом
# таким образом получим все слова что имеют одну букву Г и глассные не стоят рядом
```

```python
# Сколько существует 5-разрядных пятнадцатеричных чисел, в которых разряды, 
# делящиеся на 2 без остатка, и разряды, делящиеся на 3 без остатка, чередуются?
# Например, число 40068 подходит под описание, число 40086 — нет.
from itertools import product
c = 0
for q in product("0123456789ABCDE", repeat=5):
    q = ''.join(q)
    if q[0] != "0":
        if all([int(q[0], 15) % 2 == 0, int(q[2], 15) % 2 == 0, int(q[4], 15) % 2 == 0, int(q[1], 15) % 3 == 0,
                int(q[3], 15) % 3 == 0]):
            c += 1
        elif all([int(q[0], 15) % 3 == 0, int(q[2], 15) % 3 == 0, int(q[4], 15) % 3 == 0, int(q[1], 15) % 2 == 0,
                int(q[3], 15) % 2 == 0]):
            c += 1
print(c)
# product - все комбинации (с повоторениями) (рабоатет с исходных алфавитом)
# permutations -  все комбинации (с без повторений) (ограничение по длине алфавита)
# for i in permutations('ab',r=4) -> ['ab', 'ba']
# for i in product('ab',repeat=4): -> ['aaaa', 'aaab', 'aaba', 'aabb', 'abaa', 'abab',
# 'abba', 'abbb', 'baaa', 'baab', 'baba', 'babb', 'bbaa', 'bbab', 'bbba', 'bbbb']
```

```python
# Даша составляет 7-буквенные слова из букв БАНДЕРОЛЬ. Мягкий знак используется в каждом слове не больше одного раза.
# Остальные буквы могут быть использованы сколько угодно раз или не использоваться совсем.
# Сколько слов может составить Даша, если известно, что буква Е не может стоять рядом с согласными?
import itertools
alf = 'БАНДЕРОЛЬ'
sogl,gl = 'БНДРЛ','АОЕЬ'
c = 0
a = [''.join(p) for p in itertools.product(alf,repeat=7)]
# формируем список всех комбинаций их данного словаря длиной 7
# product - все комбинации # permutations - все перестановки
for q in a:
    if q.count('Ь') > 1:
        continue
# если Ь больше 1, то пропускаем итерацию
    f = True
    for i in range(7):
# проверяем соседей буквы Е. Если i=0, то срабатывает i<6 and q[i+1] in sogl проверяется следующая буква
# Если i=6, то срабатывает i>0 and q[i-1] in sogl проверяется предыдущая буква
# Если i между 0 и 6, то применяем оба условия
        if q[i] == 'Е' and ((i>0 and q[i-1] in sogl) or (i<6 and q[i+1] in sogl)):
            f = False
            break
    if f == True:
        c+=1
print(c)
```

```python
# Сколько существует восьмеричных шестизначных чисел, в которых все цифры различны, никакие две чётные или две нечётные 
# цифры не стоят рядом и десятичная запись которых делится на 5?
import itertools
a = [''.join(b) for b in itertools.product('01234567',repeat = 6)] # создаем список всех комбинаций длиной 6
b = [i for i in a if (i[0]!='0' and (len(set(i)) == 6) and (int(i,8)%5==0))] #строки:не начинающиеся с 0,6 различных элементов,десятичная запись кратна 5 
chet = '0246'
nechet = '1357'
c = 0
for i in b:
    f = True
    for j in range(len(i)-1):
        if (i[j] in chet and i[j+1] in chet) or (i[j] in nechet and i[j+1] in nechet): # выкидываем все строки, где два четных или нечетных числа идут подряд 
            f = False
            break
    if f == True:
        c+=1
print(c) #208
```

```python
# Рассмотрим шестнадцатеричные четырёхзначные числа. Сколько существует таких чисел, у которых:
# В записи числа присутствует две чётные и две нечётные цифры. Сумма чётных цифр равна сумме нечётных цифр. Ни одна из цифр не повторяется.
import itertools
a = [''.join(b) for b in itertools.product(alf,repeat = 4)]# создаем список всех комбинаций длиной 4
for q in a:
    if len(set(q)) == 4 and q[0]!='0': # берем строки, не начинающиеся с 0 длиною 4 различных значений
        chet = [int(i,16) for i in q if int(i,16)%2==0] # разбиваем на четные и нечетные числа
        nechet = [int(i,16) for i in q if int(i,16)%2!=0]
        if len(chet) == 2 and len(nechet) == 2 and sum(chet)==sum(nechet): # 2 четных и 2 нечетных и суммы этих значений равны
            c+=1
print(c)
```
